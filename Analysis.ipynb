```python
def getVariance(df):
  temp =  df["QUALITY"].shift() != df["QUALITY"]
  count = 0
  for i in temp:
    if i == True :
      count = count + 1
  return count
  # print(count)
```

```python
map = {
    144 : 0,
    360 : 1,
    720 : 2,
    1080 : 3
}
```

```python
def preprosessing(df) :
  name = df
  df = eval(df)
  print(name)
  df['QUALITY'] = df.apply(lambda row: row.URL.split('_')[1].split('p')[0], axis=1)
  df['TIME'] = (df['TIME'] - df['TIME'][0] )/ 1000
  df['SIZE'] =  df['SIZE'] /1000
  df['QUALITY'] = df['QUALITY'].astype('int64')
  df['Category'] = df.apply(lambda row: map[row.QUALITY], axis=1)
  df['PROTOCOL'] = name.split('_')[0]
  df['EXPERIMENT'] = name.split('_')[3]
  
  # print()
  return df , 0
```

```python
protocols = ['1','2','3','k']
downloads = ['500','1000','2000','4000']
```

```python
finalDF = pd.DataFrame(columns = ['INDEX', 'PROTOCOL', 'TIME','DOWNLOAD','TOTALREQs','VARIANCE','QUALITY'])
```

```python

def findAverage(protocol,download):
    sum = 0
    for i in range(1,4) :
      dataframe = 'http'+ protocol + '_200_' + download + '_'+ str(i)
      dataframe = eval(dataframe) 
      # dataframe,t = preprosessing(dataframe)
      sum = sum + ((dataframe['QUALITY'].sum())/(dataframe['QUALITY'].count()))
    return sum / 10

def findTime(protocol,download):
    sum = 0
    for i in range(1,4) :
      dataframe = 'http'+ protocol + '_200_' + download + '_'+ str(i)
      dataframe = eval(dataframe)
      # dataframe,t = preprosessing(dataframe)
      sum = sum + (dataframe['TIME'].iloc[-1])
    return sum/10

def variance(protocol,download):
    count = 0
    for j in range(1,4) :
      dataframe = 'http'+ protocol + '_200_' + download + '_'+ str(j)
      dataframe = eval(dataframe) 
      t = getVariance(dataframe)
      # dataframe,t = preprosessing(dataframe)
      count = count + t
    return count / 10

def totalreqs(protocol,download):
    sum = 0 
    for j in range(1,4) :
      dataframe = 'http'+ protocol + '_200_' + download + '_'+ str(j)
      dataframe = eval(dataframe)  
      # dataframe,t = preprosessing(dataframe)
      sum = sum + dataframe.shape[0]
    
    # print(protocol,download,sum/3)
    return sum / 3
    # return dataframe.shape[0]

def findQualities(finalDF) :
    index = 0
    for protocol in protocols :
      k=1
      for download in downloads:
        for j in range(1,4) :
          dataframe = 'http'+ protocol + '_200_' + download + '_'+ str(j)
          dataframe,t = preprosessing(dataframe)
        finalDF = finalDF.append({'INDEX' : index, 'PROTOCOL' : 'http'+ protocol,'TIME': findTime(protocol,download),'DOWNLOAD' : download,'TOTALREQs':totalreqs(protocol,download),'VARIANCE':variance(protocol,download), 'QUALITY':findAverage(protocol,download)}, 
                ignore_index = True)
        k=k+1
        index = index + 1

    return finalDF



```

```python
finalDF = findQualities(finalDF)
```

```python
base = 0
for index, row in percentages.iterrows():
    if index%4 == 0 :
      base = row['QUALITY']
      temp = '-'
      percentages.at[index,'%IncQualityThanHTTP1.1']=temp
    else :
      # print(row['QUALITY'])
      temp = ((row['QUALITY'] - base) / base) *100
      percentages.at[index,'%IncQualityThanHTTP1.1']=temp

newdf = percentages[['PROTOCOL','QUALITY','%IncQualityThanHTTP1.1']]
newdf['%Variance'] = percentages['VARIANCE'] / percentages['TOTALREQs'] 
newdf['DownloadSpeed'] = percentages['DOWNLOAD'] 
```
